## Latent Dim=30

```python
import numpy as np
enc5 = np.load("/Users/korayaydogan/Downloads/17.09.2022/exp5/encoded_images.npy")
enc5.shape
```

## Shannon Entropy

```python
ent = []
for i in range(enc5.shape[1]):
    pd_series = pd.Series(enc5[:,i])
    counts = pd_series.value_counts()
    entropy = scipy.stats.entropy(counts, base=2)
    ent.append(entropy)
    print(entropy)
```

```python
plt.scatter(np.arange(1, 31, 1), ent)
plt.grid()
plt.xlabel("Dimension")
plt.ylabel("Shannon_Entropy")
plt.savefig("/Users/korayaydogan/Downloads/entropy_dim.png");
```

```python
plt.plot(np.arange(1, 31, 1), ent);
```
